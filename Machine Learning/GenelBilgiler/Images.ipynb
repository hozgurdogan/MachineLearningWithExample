Model karmaşıklığı = Modelin daha detaylı tahmin edebilmesi için modelin daha detaylı hale getirmeye çalışmak.
Üstel denklemeler için Y= AX+b denekelemine Ekstra olarak x^2 eklemek örnek olarak model karamşıklığını artırmak olarak değerlendirlebilir.
Ayrıca iterasyon sayısını artırmak ve ya epoch sayısı artırmakta bir Model karmaşılığına karşılık gelir .


Eğer ki ortalama alınsa sadece karesi alınmasa - değerler ile + değerler birbirini götüreblilr bu da hatanın büyük olmasına rağmen 0 olmasına sebep olabilir.
