```python


df = pd.read_csv('../datasets/diabetes.csv')
df.head()
y=df["Outcome"]
X=df.drop(["Outcome"],axis=1)
```

```python
cat_boost_model =CatBoostClassifier(random_state=17,verbose=False).fit(X,y)
```

```python
cv_result =cross_validate(cat_boost_model,X,y,cv=5,scoring=["accuracy","f1","roc_auc"])
```

```python
cv_result['test_accuracy'].mean()
```

```python
cv_result['test_f1'].mean()
```

```python
cv_result['test_roc_auc'].mean()
```

```python
cat_boost_model.get_all_params()
```

```python
catBoost_params = {"iterations":[500,600,700,1000],
                   "learning_rate":[0.01,0.1,0.05],
                   "depth":[2,3,4,6]}
```

```python
catboost_best_grid = GridSearchCV(cat_boost_model,catBoost_params,cv=5,n_jobs=-1,verbose=True).fit(X,y)
```

```python
catboost_best_grid.best_params_
```

```python
best_params = catboost_best_grid.best_params_
# Yeni model nesnesi oluştur ve en iyi parametreleri uygula
catboost_final = CatBoostClassifier(**best_params, random_state=17).fit(X,y)
```

```python
cv_result =cross_validate(catboost_final,X,y,cv=5,scoring=["accuracy","f1","roc_auc"])
```

```python
cv_result['test_accuracy'].mean()
```

```python
cv_result['test_f1'].mean()
```

```python
cv_result['test_roc_auc'].mean()
```

```python
def plot_importance (model, features, num=len (X), save=False):
    feature_imp = pd.DataFrame({'Value': model.feature_importances_, 'Feature': features.columns})
    plt.figure(figsize=(10, 10))
    sns.set(font_scale=1)
    sns.barplot(x="Value", y="Feature", data=feature_imp.sort_values (by="Value",
                                                                      ascending=False)[0:num])
    plt.title('Features')
    plt.tight_layout()
    plt.show()
    if save:
        plt.savefig('importances.png')
```

```python
plot_importance(catboost_final,X)
```
